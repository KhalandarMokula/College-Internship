```python
def convert_to_binary(gender):
    if gender=="Male":
        return 0
    else:
        return 1
df['Bin_gender']=df['Gender'].apply(convert_to_binary)
```

```python
df
```

```python
#importing kmeans clustering algorithm from scikit-learn
from sklearn.cluster import KMeans
```

Condidering Gender and Age Columns

```python
x = df[["Age","Bin_gender"]].values
```

```python
cost = [] #wcss
for i in range(1,15):
    #instantiating the model.
    kmeans = KMeans(n_clusters=i,init='k-means++',random_state=42,n_init=20)
    #fitting  the model on the data.
    kmeans.fit(x)
    cost.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(10,5))
plt.plot(range(1,15,1),cost,marker="*",markersize=15,markerfacecolor='red')
plt.title("The Elbow Curve")
plt.xlabel("N_clusters")
plt.ylabel("WCSS COST")
```

```python
#k_clusters = 3 seems to be optimal number of clusters.
```

```python
kmeans = KMeans(n_clusters=3,init='k-means++',random_state=42,n_init=10)
```

```python
kmeans.fit(x)
```

```python
kmeans.cluster_centers_
```

```python
y_pred = kmeans.labels_
```

```python
y_pred
```

```python
plt.scatter(x[y_pred == 0, 0], x[y_pred == 0, 1], s = 100, c = 'red',alpha=0.7)
plt.scatter(x[y_pred == 1, 0], x[y_pred == 1, 1], s = 100, c = 'blue',alpha=0.7)
plt.scatter(x[y_pred == 2, 0], x[y_pred == 2, 1],s = 100, c = 'green',alpha=0.7)
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=300,c='black',alpha=1,label='Centroids',marker='*')
plt.xlabel("AGE")
plt.ylabel("Gender")
plt.title("AGE vs Gender")
```

There are equal proportion of male and female between 3 different age Groups.
